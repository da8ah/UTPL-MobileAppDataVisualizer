Covid Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('COVID-19_mobile_apps.csv', index_col='releaseDate', parse_dates=['releaseDate','latestUpdate'])
df
```

```python
df_not_indexNaN = pd.DataFrame({
    'releaseDate':df[df.index.isin(df.index.dropna())].index,
    'name':df['name'][df.index.isin(df.index.dropna())].values
}).set_index('releaseDate')

df_formated = pd.DataFrame({
    'releaseDate':[pd.to_datetime(fecha.strftime('%Y-%m-%d %H:%M:%S')) for fecha in df_not_indexNaN['name'].index],
    'name':df_not_indexNaN['name'].values
}).set_index('releaseDate').sort_index()
df_formated
```

```python
fechas = df_formated.index.dropna().drop_duplicates()
fechas = [fecha.strftime('%Y-%m') for fecha in fechas]
fechas = np.unique(fechas).tolist()
sumatorias = [df_formated['name'].loc[fecha].count() for fecha in fechas]
count_per_month = pd.DataFrame({
    'releaseDate':fechas,
    'cant':sumatorias
}).set_index('releaseDate')#.to_csv('totales.csv')

count_per_month
```
