```python
import pandas as pd
import numpy as np
pathData = './data/'
pathBase = pathData + 'base/'
```

```python
df = pd.read_csv(pathData + 'COVID-19_mobile_apps.csv', index_col='releaseDate', parse_dates=['releaseDate','latestUpdate'])
df
```

```python
def format_indexes(df):

     # Drop NaN indexes
    df_not_indexNaN = pd.DataFrame({
        'releaseDate':df[df.index.isin(df.index.dropna())].index,
        'name':df['name'][df.index.isin(df.index.dropna())].values,
        'continents':df['continents'][df.index.isin(df.index.dropna())].values,
        'geographicCoverage':df['geographicCoverage'][df.index.isin(df.index.dropna())].values
    }).set_index('releaseDate')

    # Format index with uniform datetime format
    df_formated = pd.DataFrame({
        'releaseDate':[pd.to_datetime(fecha.strftime('%Y-%m-%d %H:%M:%S')) for fecha in df_not_indexNaN['name'].index],
        'name':df_not_indexNaN['name'].values,
        'continents':df['continents'][df.index.isin(df.index.dropna())].values,
        'geographicCoverage':df['geographicCoverage'][df.index.isin(df.index.dropna())].values
    }).set_index('releaseDate').sort_index()
    
    print(df_formated)
    return df_formated

df_formated = format_indexes(df)
```

```python
def count_per_months(df_formated):

    # Getting releaseDate and Cant lists
    releaseDate = df_formated.index.dropna().drop_duplicates()
    releaseDate = [fecha.strftime('%Y-%m') for fecha in releaseDate]
    releaseDate = np.unique(releaseDate).tolist()
    cant = [df_formated['name'].loc[fecha].count() for fecha in releaseDate]
    
    # Building dataFrame with releaseDate and Cant lists
    count_per_month = pd.DataFrame({
        'releaseDate':releaseDate,
        'cant':cant
    }).set_index('releaseDate')

    # Print data to .csv
    count_per_month.to_csv(pathBase + 'statistics_count_per_month.csv')

    print(count_per_month)
    print(count_per_month['cant'].values.sum())

count_per_months(df_formated)
```

```python
# List 1
def continents(df_formated):
    continents = [continent.title() for continent in df_formated['continents'].sort_values().unique().tolist()]
    continents.insert(0, continents.pop(len(continents) - 1))
    continents = pd.DataFrame({
        'continents':continents
    }).set_index('continents')#.to_csv(pathBase + 'geographicCoverage_continents.csv')
    return continents.index.tolist()

continentsList = continents(df_formated)
```

```python
# List 2
def countries(continent):    
    geographicCoverage = df_formated[['continents','geographicCoverage']].drop_duplicates()
    geographicCoverage = df_formated['geographicCoverage'][df_formated['continents']==continent].drop_duplicates().sort_values().values
    return [countryCode for countryCode in geographicCoverage]

#[countries(continent) for continent in continentsList if continent != 'World']
countries(continentsList[1])
[code for code in countries(continentsList[1])]
```

```python
countryCodes = pd.read_csv(pathData + 'CountryCodes.csv')
countryCodes
```

```python
codes = countryCodes['alpha-3'].dropna().drop_duplicates()
len(countryCodes['name'][countryCodes['alpha-3']=='WORLD'])
```

```python
names = np.transpose([[code,countryCodes['name'][countryCodes['alpha-3']==code].item()]
for code in df_formated['geographicCoverage'].sort_values().drop_duplicates().values 
if len(countryCodes['name'][countryCodes['alpha-3']==code]) > 0])
```

```python
names = pd.DataFrame({
    'geographicCoverage':names[0,:],
    'country':names[1,:]
}).set_index('geographicCoverage')
```

```python
names
```

```python
countryCoverage = []
geographicCoverage = []
continentCoverage = []

for continent in continentsList:
    if continent != continentsList[0]:
        for code in countries(continent):
            if code != 'UAE':
                countryCoverage.append(names['country'][code].split(',')[0].split('(')[0])
                geographicCoverage.append(code)
                continentCoverage.append(continent)

pd.DataFrame({
    'country': countryCoverage,
    'geographicCoverage': geographicCoverage,
    'continent': continentCoverage
}).set_index('country').sort_index().to_csv(pathBase + 'geographicCoverage_dataframe.csv')
```

```python
contador = 0
world = []
for continent in continentsList:
    if continent != continentsList[0]:
        world.append([names['country'][code].split(',')[0].split('(')[0] for code in countries(continent) if code != 'UAE'])
        pd.DataFrame({
            continent: world[contador]
        }).set_index(continent).sort_index().to_csv(pathBase + 'geographicCoverage_' + continent.replace(' ','_').lower() + '.csv')
        contador+=1

pd.DataFrame({
    continentsList[0]:[country for fila in world for country in fila]
}).set_index(continentsList[0]).sort_index().to_csv(pathBase + 'geographicCoverage_' + continentsList[0].replace(' ','_').lower() + '.csv')
```
