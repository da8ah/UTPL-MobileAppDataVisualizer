```python
def format_indexes(df):

    # Drop NaT indexes
    df_not_indexNa = df[df.index.isin(df.index.dropna())]

    # Format index with uniform datetime format
    df_index_formatted = pd.DataFrame({
        'releaseDate':[pd.to_datetime(fecha.strftime('%Y-%m-%d')) for fecha in df_not_indexNa.index],
        'name':df_not_indexNa['name'].values,
        'continents':df_not_indexNa['continents'].values,
        'geographicCoverage':df_not_indexNa['geographicCoverage'].replace(to_replace='UAE',value='ARE').values,
        'storeName':df_not_indexNa['storeName'].values,
        'status':df_not_indexNa['status'].values
    }).dropna().set_index('releaseDate').sort_index()

    return df_index_formatted

df_index_formatted = format_indexes(df)
df_index_formatted
```

```python
def count_per_months(df_formated):

    # Getting releaseDate and Cant lists
    releaseDate = df_formated.index.dropna().drop_duplicates()
    releaseDate = [fecha.strftime('%Y-%m') for fecha in releaseDate]
    releaseDate = np.unique(releaseDate).tolist()
    cant = [df_formated['name'].loc[fecha].count() for fecha in releaseDate]
    
    # Building dataFrame with releaseDate and Cant lists
    count_per_month = pd.DataFrame({
        'releaseDate':releaseDate,
        'cant':cant
    }).set_index('releaseDate')

    # Print data to .csv
    #count_per_month.to_csv(pathBase + 'statistics_count_per_month.csv')

    print(count_per_month)
    print(count_per_month['cant'].values.sum())

count_per_months(df_index_formatted)
```

```python
def consultar_dataframe(df, continent=None, country=None, store=None, status=None, frecuency=1, movmean=1, cumulative=False):
    query = df
    if continent != None:
        query = __consultar_continent(query, continent)
    if country != None:
        query = __consultar_country(query, country)
    if store != None:
        query = __consultar_store(query, store)
    if status != None:
        query = __consultar_status(query, status)
    if frecuency >= 0 and frecuency <= 2:
        switcher = {
            0: 'D',
            1: 'M',
            2: 'Y',
        }
        if switcher.get(frecuency) =='D':
            strftime = '%Y-%m-%d'
        elif switcher.get(frecuency) == 'M':
            strftime = '%Y-%m'
        elif switcher.get(frecuency) == 'Y':
            strftime = '%Y'
        query = __aplicar_frecuency(query, switcher.get(frecuency))
    if movmean > 1 and movmean <= 15:
        query = __aplicar_movmean(query, movmean)

    if cumulative==True:
        query = __aplicar_cumulative(query)
        
    query.rename(columns = {'name':'count'}, inplace=True)
    return [query.index.strftime(strftime).tolist(),query['count'].values.tolist()]

def __consultar_continent(df, continent):
    return df[df['continents'] == continent]
def __consultar_country(df, country):
    return df[df['country'] == country]
def __consultar_store(df, store):
    return df[df['storeName'] == store]
def __consultar_status(df, status):
    return df[df['status'] == status]
def __aplicar_frecuency(df, frecuency):
    return df.resample(frecuency).count()
def __aplicar_movmean(df, movmean):
    return df.rolling(movmean).mean().dropna()
def __aplicar_cumulative(df):
    return df.cumsum()

consultar_dataframe(df,frecuency=1,movmean=3,cumulative=1)
```
